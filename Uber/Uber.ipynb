El conjunto de datos contiene información sobre la fecha, hora, latitud, longitud, y la base de cada viaje de uber que ocurrió en el mes de Septiembre de 2014 en la ciudad de Nueva York, EE. UU.

*Date/Time: La fecha y hora de la recogida de Uber

*Lat: la latitud de la recogida de Uber

*Lon: La longitud de la recogida de Uber

*Base: el código de empresa base de TLC afiliado a la recogida de Uber
*B02512 : Unter
*B02598 : Hinter
*B02617 : Weiter
*B02682 : Schmecken
*B02764 : Danach-NY

A continuacion desglozo la variable Date, en Dia y Horario

```python
uber_data['Date']=pd.to_datetime(uber_data['Date'])
```

```python
uber_data['Dates'] = pd.to_datetime(uber_data['Date']).dt.date
uber_data['Time'] = pd.to_datetime(uber_data['Date']).dt.time
```

```python
uber_data['Days'] = pd.to_datetime(uber_data['Date']).dt.day
uber_data['Mont'] = pd.to_datetime(uber_data['Date']).dt.month
uber_data['Year'] = pd.to_datetime(uber_data['Date']).dt.year
```

```python
uber_data["Weekday"] = uber_data["Date"].apply(lambda x: x.weekday())
uber_data["Hour"] = uber_data["Date"].apply(lambda x: x.hour)
```

```python
uber_data=uber_data.drop(labels=["Dates"], axis=1)
```

```python
uber_data.head(5)
```

```python
ax=sns.histplot(uber_data['Weekday'])
```

```python
ax=sns.displot(uber_data['Hour'])
```

La franja horaria con mayor frecuencia es entre las 15hs y 20hs

```python
df=uber_data.groupby(['Weekday','Hour']).apply(lambda x: len(x))
df = df.unstack()
```

```python
sns.heatmap(df,
            cmap = "Blues")
```
